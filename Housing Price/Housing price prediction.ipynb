# Model training and Evaluation

```python
#import libraries
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
X = train_df.drop(['SalePrice','Id'], axis =1)
y = train_df['SalePrice']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X, y, random_state = 0)
```

```python
# Scaling the data
from sklearn.preprocessing import RobustScaler
rscalar = RobustScaler()
x_train = rscalar.fit_transform(x_train)
x_test = rscalar.fit_transform(x_test)
test_df = rscalar.fit_transform(test_df)
```

```python
#Linear Regression 
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train,y_train)
```

```python
y_pred = linreg.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
#RidgeCV ("ridge regression with built-in cross validation of the alpha parameter")
from sklearn.linear_model import RidgeCV
ridgeCV = RidgeCV (alphas = [0.001, 0.01, 1, 10, 100], cv =5, normalize = True)
ridgeCV.fit(x_train, y_train)
```

```python
ridgeCV.alpha_
```

```python
y_pred = ridgeCV.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
#LassoCV
from sklearn.linear_model import LassoCV
lassocv = LassoCV(alphas = [0.001,0.01,1,10,100],cv = 5, normalize = True)
lassocv.fit(x_train,y_train)
```

```python
lassocv.alpha_
```

```python
y_pred = lassocv.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# XGBoost
from xgboost import XGBRegressor
xgboost = XGBRegressor()
```

```python
param_grid = dict(learning_rate = [0.1,0.01], max_depth = range(3,15), colsample_bytree = [0.4,0.5,0.6,0.7,0.8],gamma = [0,1,5])
```

```python
grid = GridSearchCV(xgboost, param_grid, cv =10, scoring = 'neg_mean_squared_error',n_jobs = -1)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
# apply best parameter to xgboost modeL
xgboost = XGBRegressor(learning_rate = 0.1 , max_depth = 3, gamma = 0, colsample_bytree = 0.5)
xgboost.fit(x_train, y_train)
```

```python
y_pred = xgboost.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

# Submission file

```python
Id = test_df['Id']
predictions = xgboost.predict(test_df.drop('Id', axis =1))

output = pd.DataFrame({'Id': Id, 'SalePrice': predictions})
output.to_csv('submission2.csv', index = False)
```
